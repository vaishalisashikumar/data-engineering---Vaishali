```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("logins").getOrCreate()
```

```python
from pyspark.sql.functions import col, trim

df = spark.read.text("logins.txt")

df = df.withColumn("value", trim(col("value")))
df.show()
```

```python
total=df.count()
print("Total Logins: ",total)
```

```python
unique=df.distinct()
unique.show()
```

```python
user_counts = df.groupBy("value").count()
user_counts.show()
```

```python
from pyspark.sql.functions import col, count, desc
top_users = user_counts.orderBy(desc("count")).limit(3)
top_users.show()
```

```python
from pyspark.sql.functions import col
active_users = user_counts.filter(col("count") > 4)
active_users.show()
```

```python
result_dict = {row["value"]: row["count"] for row in user_counts.collect()}
print(result_dict)
```
