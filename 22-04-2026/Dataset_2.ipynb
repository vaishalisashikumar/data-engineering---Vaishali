```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('employees').getOrCreate()
```

```python
df = spark.read.csv("employees.csv", header=True, inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
df.count()
```

```python
df.filter(df.department == "IT").show()
```

```python
df.filter(df.salary > 75000).show()
```

```python
from pyspark.sql.functions import avg

df.select(avg("salary")).show()
```

```python
df.orderBy(df.salary.desc()).show(1)
```

```python
df.orderBy("salary").show(1)
```

```python
df.groupBy("department").count().show()
```

```python
df.groupBy("department").agg(avg("salary")).show()
```

```python
df.groupBy("city").count().show()
```

```python
df.orderBy(df.salary.desc()).show(5)
```

```python
df.filter((df.city == "Hyderabad")&(df.salary>70000)).show()
```
