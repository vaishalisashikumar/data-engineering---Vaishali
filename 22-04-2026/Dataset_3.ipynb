```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('sales').getOrCreate()
```

```python
df_sales = spark.read.csv("sales.csv", header=True, inferSchema=True)
df_sales.show()
```

```python
from pyspark.sql.functions import col

df_sales = df_sales.withColumn("revenue", col("quantity") * col("price"))
df_sales.show()
```

```python
from pyspark.sql.functions import sum

df_sales.select(sum("revenue")).show()
```

```python
df_sales.groupBy("product") \
    .agg(sum("revenue").alias("total_revenue")) \
    .show()
```

```python
df_sales.groupBy("product") \
    .agg(sum("quantity").alias("total_qty")) \
    .show()
```

```python
df_sales.groupBy("product") \
    .agg(sum("quantity").alias("total_qty")) \
    .show()
```

```python
df_sales.groupBy("product") \
    .agg(sum("quantity").alias("total_qty")) \
    .orderBy(col("total_qty").desc()) \
    .show(1)
```

```python
df_sales.groupBy("emp_id") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc()) \
    .show(1)
```

```python
from pyspark.sql.functions import avg

df_sales.select(avg("revenue")).show()
```

```python
df_sales.groupBy("product") \
    .agg(sum("revenue").alias("total_revenue")) \
    .filter(col("total_revenue") > 100000) \
    .show()
```
