## Section 10

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
w = Window.partitionBy('city').orderBy(col('balance').desc())
customers.join(accounts,'customer_id').withColumn('rank', rank().over(w)).show()
```

```python
from pyspark.sql.functions import row_number
w = Window.partitionBy('city').orderBy(col('balance').desc())
df = customers.join(accounts,'customer_id').withColumn('rn', row_number().over(w))
df.filter(col('rn')==1).show()
```

```python
from pyspark.sql.functions import dense_rank
w = Window.partitionBy('txn_type').orderBy(col('amount').desc())
transactions.withColumn('dense_rank', dense_rank().over(w)).show()
```

```python
w = Window.partitionBy('account_id').orderBy(col('amount').desc())
df = transactions.withColumn('rn', row_number().over(w))
df.filter(col('rn') <= 2).show()
```

```python
from pyspark.sql.functions import sum as spark_sum
w = Window.partitionBy('account_id').orderBy('txn_date').rowsBetween(Window.unboundedPreceding, 0)
transactions.withColumn('running_total', spark_sum('amount').over(w)).show()
```

```python
from pyspark.sql.functions import rank
branch_bal = accounts.groupBy('branch').agg(sum('balance').alias('total_balance'))
w = Window.orderBy(col('total_balance').desc())
branch_bal.withColumn('rank', rank().over(w)).show()
```

```python
city_bal = customers.join(accounts,'customer_id').groupBy('city').agg(sum('balance').alias('total_balance'))
w = Window.orderBy(col('total_balance').desc())
city_bal.withColumn('rank', rank().over(w)).show()
```

```python
cust_txn = customers.join(accounts,'customer_id').join(transactions,'account_id')\
.groupBy('name').agg(sum('amount').alias('total_txn'))
w = Window.orderBy(col('total_txn').desc())
cust_txn.withColumn('rank', rank().over(w)).show()
```

```python
df = customers.join(accounts,'customer_id').join(branches,'branch')
w = Window.partitionBy('region').orderBy(col('balance').desc())
df.withColumn('rn', row_number().over(w)).filter(col('rn')==1)\
  .select('region','name','balance').show()
```
