```python
from pyspark.sql.functions import when, col

patients.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when(col("age") < 40, "Adult")
    .otherwise("Senior")
).show()
```

```python
bills.withColumn(
    "bill_category",
    when(col("bill_amount") >= 1500, "High")
    .when(col("bill_amount") >= 1000, "Medium")
    .otherwise("Low")
).show()
```

```python
doctors.withColumn(
    "fee_category",
    when(col("consultation_fee") >= 1500, "Premium")
    .when(col("consultation_fee") >= 1000, "Standard")
    .otherwise("Basic")
).show()
```

```python
appointments.withColumn(
    "priority",
    when(col("status") == "Pending", "High")
    .when(col("status") == "Completed", "Normal")
    .otherwise("Low")
).show()
```

```python
patients.withColumn(
    "city_zone",
    when(col("city").isin("Hyderabad", "Bangalore", "Chennai"), "South")
    .otherwise("Metro")
).show()
```

```python
spark.sql("""
SELECT p.name, a.appointment_date, a.status
FROM patients p
JOIN appointments a
ON p.patient_id = a.patient_id
""").show()

spark.sql("""
SELECT d.doctor_name, a.appointment_date, a.status
FROM doctors d
JOIN appointments a
ON d.doctor_id = a.doctor_id
""").show()

spark.sql("""
SELECT a.appointment_id, a.status, b.bill_amount, b.payment_status
FROM appointments a
JOIN bills b
ON a.appointment_id = b.appointment_id
""").show()

spark.sql("""
SELECT payment_mode, SUM(bill_amount)
FROM bills
GROUP BY payment_mode
""").show()

spark.sql("""
SELECT doctor_name, consultation_fee,
RANK() OVER (ORDER BY consultation_fee DESC) as rank
FROM doctors
""").show()

spark.sql("""
SELECT p.name, SUM(b.bill_amount) as total_bill
FROM patients p
JOIN appointments a ON p.patient_id = a.patient_id
JOIN bills b ON a.appointment_id = b.appointment_id
GROUP BY p.name
ORDER BY total_bill DESC
LIMIT 1
""").show()
```
