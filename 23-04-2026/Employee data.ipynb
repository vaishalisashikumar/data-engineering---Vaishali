```python
from pyspark.sql import SparkSession
session=SparkSession.builder.appName("Dataset-2").getOrCreate()
sc=session.sparkContext
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),

("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

```python
emp_names = emp_rdd.map(lambda x: x[0])
emp_names.collect()
```

```python
high_salary = emp_rdd.filter(lambda x: x[2] > 70000)
high_salary.collect()
```

```python
departments = emp_rdd.map(lambda x: x[1]).distinct()
departments.collect()
```

```python
salary_by_dept = emp_rdd \
  .map(lambda x: (x[1], x[2])) \
  .reduceByKey(lambda a, b: a + b)
salary_by_dept.collect()
```

```python
avg_salary = emp_rdd \
  .map(lambda x: (x[1], (x[2], 1))) \
  .reduceByKey(lambda a, b: (a[0]+b[0], a[1]+b[1])) \
  .map(lambda x: (x[0], x[1][0] / x[1][1]))
avg_salary.collect()
```

```python
top_emp = emp_rdd \
  .sortBy(lambda x: x[2], ascending=False) \
  .take(1)
print(top_emp)
```

```python
emp_count = emp_rdd \
  .map(lambda x: (x[1], 1)) \
  .reduceByKey(lambda a, b: a + b)
emp_count.collect()
```

```python
sorted_emp = emp_rdd.sortBy(lambda x: x[2], ascending=False)
sorted_emp.collect()
```

```python
top3 = emp_rdd \
  .sortBy(lambda x: x[2], ascending=False) \
  .take(3)
print(top3)
```

```python
dept_name = emp_rdd.map(lambda x: (x[1], x[0]))
dept_name.collect()
```

```python
grouped = emp_rdd \
  .map(lambda x: (x[1], x[0])) \
  .groupByKey() \
  .map(lambda x: (x[0], list(x[1])))
grouped.collect()
```

```python
top_dept = emp_rdd \
  .map(lambda x: (x[1], x[2])) \
  .reduceByKey(lambda a, b: a + b) \
  .sortBy(lambda x: x[1], ascending=False) \
  .take(1)
print(top_dept)
```

```python
dept_summary = emp_rdd \
  .map(lambda x: (x[1], (1, x[2]))) \
  .reduceByKey(lambda a, b: (a[0]+b[0], a[1]+b[1]))
dept_summary.collect()
```
