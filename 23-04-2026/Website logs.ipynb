```python
from pyspark.sql import SparkSession
session=SparkSession.builder.appName("Dataset-1").getOrCreate()
sc=session.sparkContext
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]

logs_rdd = sc.parallelize(logs)
logs_rdd.collect()
```

```python
logs_rdd.count()
```

```python
logs_rdd.take(5)
```

```python
names_rdd = logs_rdd.map(lambda x: x.split()[0])
names_rdd.collect()
```

```python
actions_rdd = logs_rdd.map(lambda x: x.split()[1])
actions_rdd.collect()
```

```python
unique_users = logs_rdd.map(lambda x: x.split()[0]).distinct()
unique_users.collect()
```

```python
login_count = logs_rdd.filter(lambda x: x.split()[1] == "login").count()
print(login_count)
```

```python
purchases = logs_rdd.filter(lambda x: x.split()[1] == "purchase")
purchases.collect()
```

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
```

```python
words_rdd = sentences.flatMap(lambda x: x.split())
words_rdd.collect()
```

```python
distinct_tech = sentences.flatMap(lambda x: x.split()).distinct()
distinct_tech.collect()
```

```python
word_counts = sentences \
  .flatMap(lambda x: x.split()) \
  .map(lambda word: (word, 1)) \
  .reduceByKey(lambda a, b: a + b)
word_counts.collect()
```
