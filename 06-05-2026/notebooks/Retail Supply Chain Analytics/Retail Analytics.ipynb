#part - 8
Merge / Upsert

```python
orders_df.write.format("delta").mode("overwrite").saveAsTable("orders_target")
spark.table("orders_target").show()
daily_orders_data=[
(321,114,204,"2024-04-11",3,"Delivered"),
(322,118,210,"2024-04-11",2,"Delivered"),
(304,114,208,"2024-04-02",5,"Delivered"),
(319,112,208,"2024-04-10",2,"Delivered"),
(323,120,210,"2024-04-12",1,"Pending")
]
daily_orders_columns=[
"order_id",
"product_id",
"supplier_id",
"order_date",
"quantity",
"order_status"
]
daily_orders_df=spark.createDataFrame(daily_orders_data,daily_orders_columns)
daily_orders_df=daily_orders_df.withColumn("order_date",to_date(col("order_date")))
daily_orders_df.createOrReplaceTempView("daily_orders_updates")
```

```python
spark.sql("""
MERGE INTO orders_target t
USING daily_orders_updates s
ON t.order_id=s.order_id
WHEN MATCHED THEN
UPDATE SET
t.product_id=s.product_id,
t.supplier_id=s.supplier_id,
t.order_date=s.order_date,
t.quantity=s.quantity,
t.order_status=s.order_status
WHEN NOT MATCHED THEN
INSERT (
order_id,
product_id,
supplier_id,
order_date,
quantity,
order_status
)
VALUES (
s.order_id,
s.product_id,
s.supplier_id,
s.order_date,
s.quantity,
s.order_status
)
""")
spark.table("orders_target").filter(col("order_id")==304).show()
spark.table("orders_target").filter(col("order_id")==319).show()
```

```python
spark.table("orders_target").filter(col("order_id")==321).show()
spark.table("orders_target").filter(col("order_id")==322).show()
spark.table("orders_target").filter(col("order_id")==323).show()
spark.table("orders_target").filter(col("order_id").isin(304,319)).show()
```

```python
spark.table("orders_target").filter(col("order_id").isin(321,322,323)).show()
spark.sql("DESCRIBE HISTORY orders_target").show()
print("MERGE is important because it supports UPSERT operations by updating existing records and inserting new records efficiently in Delta Lake.")
```
