```python
daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]
daily_visits_columns = [
"visit_id",
"patient_id",
"doctor_id",
"visit_date",
"visit_status",
"tests_count"
]
daily_visits_df = spark.createDataFrame(
daily_visits_data,
daily_visits_columns
)
daily_visits_df = daily_visits_df.withColumn(
"visit_date",
to_date(col("visit_date"),"yyyy-MM-dd")
)
daily_visits_df.createOrReplaceTempView(
"daily_updates"
)
daily_visits_df.show()
```

```python
delta_table = DeltaTable.forName(
spark,
"visits_target"
)
delta_table.alias("target").merge(
daily_visits_df.alias("source"),
"target.visit_id = source.visit_id"
).whenMatchedUpdate(set = {
"patient_id":"source.patient_id",
"doctor_id":"source.doctor_id",
"visit_date":"source.visit_date",
"visit_status":"source.visit_status",
"tests_count":"source.tests_count"
}).whenNotMatchedInsert(values = {
"visit_id":"source.visit_id",
"patient_id":"source.patient_id",
"doctor_id":"source.doctor_id",
"visit_date":"source.visit_date",
"visit_status":"source.visit_status",
"tests_count":"source.tests_count"
}).execute()
spark.table("visits_target").show()
```
