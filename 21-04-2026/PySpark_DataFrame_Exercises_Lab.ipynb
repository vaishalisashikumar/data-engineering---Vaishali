```python
from pyspark.sql.functions import col, lit

# ex31
employees_df = employees_df.withColumn("bonus", col("salary") * 0.1)

# ex32
employees_df = employees_df.withColumn("company", lit("BotCampus"))

employees_df.show()
```

PART-10

```python
from pyspark.sql.functions import col
# ex33
employees_df.filter(col("department").isNull()).show()
# ex34
employees_df = employees_df.fillna({"department": "Unknown"})
employees_df.show()
# ex35
employees_df.dropna().show()
```

PART-11

```python
from pyspark.sql.functions import upper, length, col

# ex36
employees_df.select(upper("name").alias("name_upper")).show()

# ex37
employees_df.select("name", length("name").alias("name_length")).show()

# ex38
employees_df = employees_df.withColumn("salary_in_lakhs", col("salary") / 100000)

employees_df.show()
```

Part-12

```python
from pyspark.sql.functions import when, col

# ex39
employees_df = employees_df.withColumn(
    "salary_grade",
    when(col("salary") >= 75000, "High")
    .when((col("salary") >= 60000) & (col("salary") < 75000), "Medium")
    .otherwise("Low")
)

employees_df.show()
```

Part-13

```python
from pyspark.sql.functions import to_date, year, month

# ex40
employees_df = employees_df.withColumn("joining_date", to_date("joining_date"))

# ex41
employees_df = employees_df.withColumn("joining_year", year("joining_date")) \
                           .withColumn("joining_month", month("joining_date"))

employees_df.show()
```

```python
from pyspark.sql import Row

# ex42
new_data = [
    (8, "Ravi", "IT", 65000, "2023-05-01"),
    (9, "Anu", "HR", 62000, "2023-06-10")
]

new_df = spark.createDataFrame(new_data, employees_df.columns[:5])

union_df = employees_df.select("emp_id", "name", "department", "salary", "joining_date") \
    .union(new_df)

union_df.show()

# ex43
union_df2 = employees_df.select("emp_id", "name", "department", "salary", "joining_date") \
    .unionByName(new_df)

union_df2.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, row_number

window_spec = Window.partitionBy("department").orderBy(col("salary").desc())

# ex44
employees_df.withColumn("rank", rank().over(window_spec)).show()

# ex45
employees_df.withColumn("row_number", row_number().over(window_spec)).show()
```

```python
from pyspark.sql.functions import sum

# ex49 — Total sales
sales_df.select(sum("amount")).show()

# ex50 — Total sales per employee
sales_df.groupBy("emp_id").sum("amount").show()

# ex51 — Best performing employee
sales_df.groupBy("emp_id").sum("amount") \
    .orderBy(col("sum(amount)").desc()) \
    .show(1)

# ex52 — Total sales per product
sales_df.groupBy("product").sum("amount").show()
```

part-18

```python
from pyspark.sql.functions import when

# ex53 — Create sale_category
sales_df = sales_df.withColumn(
    "sale_category",
    when(col("amount") > 50000, "Premium")
    .when((col("amount") >= 20000) & (col("amount") <= 50000), "Medium")
    .otherwise("Small")
)

sales_df.show()
```

part-20

```python
import ipywidgets as widgets
from IPython.display import display

# ex56 — Dropdown
dept_widget = widgets.Dropdown(
    options=employees_df.select("department").distinct().rdd.flatMap(lambda x: x).collect(),
    description='Dept:'
)

# ex57 — Slider
salary_widget = widgets.IntSlider(min=40000, max=80000, step=1000, description='Salary')

display(dept_widget, salary_widget)

# ex58 — Filter based on widgets
filtered_df = employees_df.filter(
    (col("department") == dept_widget.value) & (col("salary") >= salary_widget.value)
)

filtered_df.show()
```

```python
# ==============================
# FINAL MINI PROJECT
# ==============================

from pyspark.sql.functions import col, avg, count, sum

# ------------------------------
# Top Employee (highest salary)
# ------------------------------
top_employee = employees_df.orderBy(col("salary").desc()).select("name", "salary").first()

# ------------------------------
# Average Salary
# ------------------------------
avg_salary = employees_df.select(avg("salary")).first()[0]

# ------------------------------
# Department Employee Count
# ------------------------------
dept_count = employees_df.groupBy("department").count().collect()

# ------------------------------
# Total Sales
# ------------------------------
total_sales = sales_df.select(sum("amount")).first()[0]

# ------------------------------
# Best Sales Employee
# ------------------------------
best_sales = sales_df.groupBy("emp_id").sum("amount") \
    .orderBy(col("sum(amount)").desc()) \
    .first()

# ------------------------------
# Convert emp_id → name
# ------------------------------
best_emp_name = employees_df.filter(col("emp_id") == best_sales["emp_id"]) \
    .select("name").first()[0]

# ------------------------------
# Create Report Text
# ------------------------------
report = ""

report += "===== FINAL REPORT =====\n\n"

report += f"Top Employee: {top_employee['name']} (Salary: {top_employee['salary']})\n\n"

report += f"Average Salary: {avg_salary}\n\n"

report += "Department Employee Count:\n"
for row in dept_count:
    report += f"{row['department']} → {row['count']}\n"

report += "\n"

report += f"Total Sales: {total_sales}\n\n"

report += f"Best Sales Employee: {best_emp_name} (Sales: {best_sales['sum(amount)']})\n"

# ------------------------------
# Save to file
# ------------------------------
with open("final_report.txt", "w") as f:
    f.write(report)

print("Report saved as final_report.txt")
```
