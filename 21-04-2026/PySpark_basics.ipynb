```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Python Basics").getOrCreate()
print("Spark Session Created")
```

```python
print("version of sparkL: ",spark.version)
```

```python
#dataframe
data = [
("Rahul", 25, "Hyderabad"),
("Sneha", 28, "Bangalore"),
("Arjun", 24, "Chennai"),
("Priya", 27, "Hyderabad")

]
columns=["Name","Age","City"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.printSchema()
df.select("Name").show()
df.select("Name","Age").show()
```

```python
df.filter(df.Age>25).show()
```

```python
df.groupBy("City").count().show()

df.orderBy("Age").show()

df.orderBy(df.Age.desc()).show()
```
