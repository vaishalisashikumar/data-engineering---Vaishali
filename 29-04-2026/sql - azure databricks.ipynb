# Conditional queries

```sql
select * from employees
where salary > (select avg_salary from (select avg(salary) as avg_salary from employees) as sub);
```

```sql
SELECT * FROM employees e
WHERE experience > (
    SELECT AVG(experience) 
    FROM employees 
    WHERE department = e.department
);
```

```sql
select * from employees
where salary<70000
```

```sql
select * from employees
where department='HR'
```

# JOINS

```sql
select e.department,p.project_name,p.budget
from employees e 
join projects p
on e.department=p.department;
```

```sql
select e.emp_name,e.department,p.project_name
from employees e 
join projects p
on e.department=p.department;
```

```sql
select p.department, sum(p.budget) as total_budget
from projects p
group by p.department;
```

```sql
select distinct e.* 
from employees e
where e.department in (select department from projects);
```

```sql
select p.project_name, count(e.emp_id) as employee_count
from projects p
join employees e
on p.department=e.department
group by p.project_name;
```

# Final mini capstone

```sql
create table employee_salary_analysis as
select 
    emp_name,
    city,
    department,
    salary,
    experience,
    case 
        when salary >= 95000 then 'high'
        when salary >= 80000 then 'medium'
        else 'low'
    end as salary_category,
    avg(salary) over (partition by department) as department_avg_salary
from employees;
```
