## Section 8 — *Window* *Functions*

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
w = Window.orderBy(col("price").desc())
products.withColumn("rank", rank().over(w)).show()
```

```python
w = Window.partitionBy("category").orderBy(col("price").desc())
products.withColumn("rank", rank().over(w)).show()
```

```python
from pyspark.sql.functions import row_number
w = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(w)).show()
```

```python
cust_rev = orders.join(order_items, "order_id") \
  .join(products, "product_id") \
  .join(customers, "customer_id") \
  .withColumn("revenue", col("quantity") * col("price")) \
  .groupBy("name").agg(sum("revenue").alias("revenue"))

w = Window.orderBy(col("revenue").desc())
cust_rev.withColumn("rank", rank().over(w)).show()
```

```python
from pyspark.sql.functions import dense_rank
w = Window.orderBy(col("revenue").desc())
cust_rev.withColumn("dense_rank", dense_rank().over(w)).show()
```

```python
cust_city_rev = orders.join(order_items, "order_id") \
  .join(products, "product_id") \
  .join(customers, "customer_id") \
  .withColumn("revenue", col("quantity") * col("price")) \
  .groupBy("name", "city").agg(sum("revenue").alias("revenue"))

w = Window.partitionBy("city").orderBy(col("revenue").desc())
cust_city_rev.withColumn("rank", rank().over(w)).filter(col("rank") == 1).show()
```

```python
prod_rev = order_items.join(products, "product_id") \
  .withColumn("revenue", col("quantity") * col("price")) \
  .groupBy("product_name", "category").agg(sum("revenue").alias("revenue"))

w = Window.partitionBy("category").orderBy(col("revenue").desc())
prod_rev.withColumn("rank", rank().over(w)).filter(col("rank") == 1).show()
```

```python
from pyspark.sql.functions import to_date
order_rev = orders.join(payments, "order_id") \
  .select("order_date", "amount")

w = Window.orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)
order_rev.withColumn("running_total", sum("amount").over(w)).show()
```

```python
prod_item_rev = order_items.join(products, "product_id") \
  .withColumn("revenue", col("quantity") * col("price"))

w = Window.partitionBy("product_name").orderBy("order_id") \
  .rowsBetween(Window.unboundedPreceding, Window.currentRow)
prod_item_rev.withColumn("running_total", sum("revenue").over(w)).show()
```

```python
w = Window.orderBy(col("amount").desc())
payments.withColumn("rank", rank().over(w)).show()
```

## Section 10 — Spark SQL

```python
customers.createOrReplaceTempView("customers")
products.createOrReplaceTempView("products")
orders.createOrReplaceTempView("orders")
order_items.createOrReplaceTempView("order_items")
payments.createOrReplaceTempView("payments")

spark.sql("SELECT * FROM customers").show()
```

```python
spark.sql("SELECT * FROM customers WHERE age > 30").show()
```

```python
spark.sql("SELECT city, COUNT(*) as count FROM customers GROUP BY city").show()
```

```python
spark.sql("SELECT city, AVG(age) as avg_age FROM customers GROUP BY city").show()
```

```python
spark.sql("""
  SELECT c.name, o.order_id, o.status
  FROM customers c
  JOIN orders o ON c.customer_id = o.customer_id
""").show()
```

```python
spark.sql("""
  SELECT c.name, COUNT(o.order_id) as total_orders
  FROM customers c
  JOIN orders o ON c.customer_id = o.customer_id
  GROUP BY c.name
""").show()
```

```python
spark.sql("""
  SELECT oi.order_id, p.product_name, oi.quantity, p.price
  FROM order_items oi
  JOIN products p ON oi.product_id = p.product_id
""").show()
```

```python
spark.sql("""
  SELECT oi.order_id, SUM(oi.quantity * p.price) as revenue
  FROM order_items oi
  JOIN products p ON oi.product_id = p.product_id
  GROUP BY oi.order_id
""").show()
```

```python
spark.sql("""
  SELECT c.name, SUM(oi.quantity * p.price) as revenue
  FROM customers c
  JOIN orders o ON c.customer_id = o.customer_id
  JOIN order_items oi ON o.order_id = oi.order_id
  JOIN products p ON oi.product_id = p.product_id
  GROUP BY c.name
  ORDER BY revenue DESC
  LIMIT 3
""").show()
```
